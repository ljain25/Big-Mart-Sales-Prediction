**Preprocessing - Removing duplicates and null values**

```python
# check for duplicates
big_mart_df.duplicated().sum()
```

There are no duplicates in the df

Replacing the null values of Item_Weight

```python
# check for null values
big_mart_df.isnull().sum()
```

```python
# Replacing the null values in Item_weight
# First we need to plot the distribution plot
fig = px.histogram(big_mart_df, x = "Item_Weight")
fig.show()
```

```python
# since we don't see any interval which has large no. of datapoints, we can replace the null values by mean
avg = big_mart_df["Item_Weight"].mean()
avg
```

```python
big_mart_df["Item_Weight"].fillna(avg, inplace = True)
```

```python
big_mart_df["Item_Weight"].isnull().sum()
# the values replaced successfully
```

Replacing null values in "Outlet_Size" with mode

```python
# The Outlet_size depends on the Outlet_type. Grocery store would be small. But supermarkets will be larger in isze
mode_outlet_size = big_mart_df.pivot_table(values = "Outlet_Size", columns = "Outlet_Type", aggfunc = (lambda x: x.mode()[0]))
mode_outlet_size
```

The outlet_size null values would be filled corresponding to the type

```python
missing_values = big_mart_df["Outlet_Size"].isnull()
```

```python
print(missing_values)
```

```python
big_mart_df.loc[missing_values, "Outlet_Size"] = big_mart_df.loc[missing_values, "Outlet_Type"].apply(lambda x: mode_outlet_size)
```
